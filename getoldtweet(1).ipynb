```python
df = pd.read_csv('movielist_startenddate.csv',encoding = 'ISO-8859-1')
```

```python
df['Startdate'].head()
```

```python
df['Enddate'].head()
```

```python
for i,dat in enumerate(df.Startdate):
    li = dat.split('-')
    li
```

```python
for i,dat in enumerate(df.Startdate):
    df.Startdate[i] =  dat.replace('/','-')
    df.Enddate[i] =  df.Enddate[i].replace('/','-')
df
```

```python
res = []
for i in range(df.shape[0]):
    name = df['movie_title'][i]
    startdate = df['Startdate'][i]
    enddate = df['Enddate'][i]
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(str(name)).setSince(str(startdate)).setUntil(str(enddate)).setMaxTweets(100)
    tweet = got.manager.TweetManager.getTweets(tweetCriteria)
    tweet_list = [j.text for j in tweet]
    tweet_list[:] = [''.join(tweet_list[:])] 
    res.append(str(tweet_list))    
```

```python
res
```

```python
print(res[0])
```

```python
df3=pd.DataFrame(res)
```

```python
df3['Movie'] = df['movie_title']
df3['Startdate'] = df['Startdate']
df3['Enddate'] = df['Enddate']
```
