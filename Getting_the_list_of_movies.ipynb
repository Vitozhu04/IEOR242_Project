## 1.Getting the list from the IMDB kaggle datset

```python
import pandas as pd 
import numpy as np
from datetime import datetime

def str_to_datetime(datelist,ymd_type):
    datelist = datelist.dropna()
    for i,dat in enumerate(datelist):
        datelist[i] = datetime.strptime(dat, ymd_type)
    return datelist
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')

#See the objects with missing value
movies[movies.isnull().values == True]
```

```python
movies.release_date = str_to_datetime(movies.release_date,'%Y-%m-%d')

movies_title = []
for i in range(len(movies.release_date)):
    if type(movies['release_date'][i]) != datetime:
        print(i)
        movies = movies.drop([i])
        continue
    if movies['release_date'][i].year < 2014:
        movies_title.append(movies.original_title)
        movies = movies.drop([i])

print('The number of the movie after 2014(included) is:',len(movies))

movies.to_csv('movies_after2014_IMDB.csv')
```

## 2.Getting the movie list from the wikipedia

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>")) 
from __future__ import division, print_function
import requests 
import bs4 as bs 
import re
```

```python
def get_list_of_movie(html,last_movie_name):
    source = requests.get(html)

    soup = bs.BeautifulSoup(source.content, features='html.parser') 

    links = soup.find_all('i')
    links2 = soup.find_all('td')

    # 1.getting the duplicate ones named as Sequel to ...
    sequel_to_link = []
    for li2 in links2:
        if len(re.findall(r'Sequel to.*',str(li2))) > 0:
            sequel_to_link.append(re.findall(r'Sequel to.*',str(li2))[0])

    sequel_i_link = []
    for li in sequel_to_link:
        sequel_i_link.append(re.findall(r'<i>.*</i>',li))
        #print(re.findall(r'<i>.*</i>',li))

    title2 = []
    for li in sequel_i_link:
        if len(re.findall(r'title=".*"',str(li))) >= 1:
            title2.append(re.findall(r'title=".*"',str(li))[0][7:-1])
        
    # 2.getting all the titles
    title = []
    for li in links:
        if len(re.findall(r'title=".*"',str(li))) >= 1:
            title.append(re.findall(r'title=".*"',str(li))[0][7:-1])

    titlelist = [i for i in title if i not in title2]

    # 3.deleting the () in the title
    for i,title in enumerate(titlelist):
        titlelist[i] = title.split('(')[0]
        if titlelist[i][-1] == " ":
            titlelist[i] = titlelist[i][:-1]
        if titlelist[i] == last_movie_name:
            break
    titlelist = list(set(titlelist[:i+1]))
    print(len(titlelist))
    return np.array((titlelist))
```

```python
# Using the link on the wiki and the last movie on the list 
# to get the whole list of American movie of that year
titlelist_2014 = get_list_of_movie("https://en.wikipedia.org/wiki/List_of_American_films_of_2014","Unbroken")
titlelist_2015 = get_list_of_movie("https://en.wikipedia.org/wiki/List_of_American_films_of_2015","The Revenant")
titlelist_2016 = get_list_of_movie("https://en.wikipedia.org/wiki/List_of_American_films_of_2016","Rings")
titlelist_2017 = get_list_of_movie("https://en.wikipedia.org/wiki/List_of_American_films_of_2017","Phantom Thread")
titlelist_2018 = get_list_of_movie("https://en.wikipedia.org/wiki/List_of_American_films_of_2018","Black Mirror: Bandersnatch")

# combine all the list with their year
titlelist_2014 = pd.DataFrame(data=titlelist_2014,columns=['Name'])
titlelist_2014['Year'] = 2014
titlelist_2015 = pd.DataFrame(data=titlelist_2015,columns=['Name'])
titlelist_2015['Year'] = 2015
titlelist_2016 = pd.DataFrame(data=titlelist_2016,columns=['Name'])
titlelist_2016['Year'] = 2016
titlelist_2017 = pd.DataFrame(data=titlelist_2017,columns=['Name'])
titlelist_2017['Year'] = 2017
titlelist_2018 = pd.DataFrame(data=titlelist_2018,columns=['Name'])
titlelist_2018['Year'] = 2018

# change the index
titlelist = pd.concat([titlelist_2014,titlelist_2015,titlelist_2016,titlelist_2017,titlelist_2018],axis=0)
titlelist.index = range(len(titlelist))

titlelist.to_csv('movies_after2014_wiki.csv')
```
